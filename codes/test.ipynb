```python
# library
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from matplotlib import cm
import numpy as np

scores = pd.read_csv('/home/guillaume/Documents/Env_DMLA/results/005_AU_R/IR OD/best_score.csv')
df=scores[['patch_size','sigma','f1_score']]
df.columns=["X","Y","Z"]

# Make the plot
fig = plt.figure()
ax = Axes3D(fig)
surf = ax.plot_trisurf(df['Y'], df['X'], df['Z'], cmap=plt.cm.viridis, linewidth=0.1)
fig.colorbar(surf, shrink=0.5, aspect=5)
# ax.plot_trisurf(df['Y'], df['X'], df['Z'], cmap=plt.cm.viridis, linewidth=0.2)
ax.set_xlabel('Sigma')
ax.set_ylabel('Patch size')
ax.set_title('F1-Score')
# ax.view_init(10, -80)
plt.savefig('sigmapatchf1.png')
plt.show()


# Rotate it
# ax.view_init(45, 30)
# plt.show()
```
