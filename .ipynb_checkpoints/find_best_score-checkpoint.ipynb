```python
import os
import pandas as pd
import numpy as np
import re

patch_sizes = [5,7,9,11,13,15,17]
patient ='005_AU_R/IR OD/'
racine = '/home/guillaume/Documents/Env_DMLA/results/'+patient
results = []

for patch_size in patch_sizes:
    path = racine+'patch_size_'+str(patch_size)+'/weighted/'
    for subdir, dirs, files in os.walk(path):
        if 'score' in subdir:
            score = pd.read_csv(subdir+'/scores.csv')
            sigma = subdir[-8:]
            sigma = int(re.sub("\D", "", sigma))
            results.append(np.append([int(patch_size),int(sigma)],np.array(score)[1][1:]))
            columns=  score.columns[1:]

columns = np.append(['patch_size','sigma'],columns)
results = pd.DataFrame(results)
results.columns=columns
results = results.sort_values('f1_score',ascending=False)
results.reset_index(drop=True,inplace=True)
results.to_csv(racine+'best_score.csv',header=True)
```
